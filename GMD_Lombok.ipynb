```python
# Mengganti nilai missing value pada kolom tertentu dengan forward fill (ffill)
kolom_ffill = ['NAMA KABUPATEN', 'NAMA KECAMATAN', 'NAMA DESA']
data3[kolom_ffill] = data3[kolom_ffill].fillna(method='ffill')
```

###Contoh

Berdasarkan hasil pengolahan dan analisis diatas,
1. Pendidikan:

  Potensi:
  1. Tersedia lahan yang cukup luas
  2. Peluang Kerja yang tinggi untuk pariwisata setempat
  3. Adanya lulusan sarjana bahasa asing yang bisa diberdayakan

  Masalah:
  1. Terbatasnya sarana baca dan perpustakaan di desa dan sekolah
  2. Angka Melek Huruf yang cukup rendah karena tingat buta huruf yang tinggi
  3. Angka Partisipasi Kasar dan Angka Pastisipasi Murni yang cukup rendah dibandingkan angka nasional terutama untuk SMP/SMA
  4. Banyak siswa SMP/SMA yang drop-out

  Program:
  1. Gerakan Literasi Baca Buku
  2. Ajar Paket B/C
  3. Mari berbahasa Inggris

  Tujuan:
  1. Gerakan Literasi Baca Buku, yaitu untuk peningkatan kesadaran masyarakat terhadap literasi buku bacaan, membantu menyadarkan warga akan mindset yang terbelakang
  2. Ajar Paket B/C, yaitu pembekalan untuk masyarakat yang drop out dari SMP/SMA agar dapat meneruskan pendidikan ke tigkat yang lebih lanjut, terutama yang drop out SMA bisa kuliah dengan pembekalan beasiswanya
  3. Mari Berbahasa Inggris, meningkatkan kemampuan masyarakat terutama pemuda setempat dalam berbahasa inggris sehingga mampu berinteraksi dengan orang asing sebagai pemandu wisata
  
  Implementasi:
  1. Gerakan Literasi Baca Buku: Gerakan donasi buku (Gedebug), gerakan sosialisasi baca buku, gerakan membangun pojok baca desa, gerakan pohon cita-cita.
  2. Ajar Paket B/C: Sosialisasi pentingnya belajar kepada masyarakat bisa dibarengi dengan baca buku, pendampingan belajar masyarakat putus sekolah dari jenjang SMP/SMA, kerjasama program beasiswa dengan lembaga lain.
  3. Mari Berbahasa Inggris: Pendampingan untuk pemuda daerah setempat untuk belajar bahasa inggris, penyediaan akses literasi bahasa inggris via handphone, penyediaan buku bacaan berbahasa inggris.

  Teknis Acara: Mempertimbangkan banyak hal dari divisi lain juga
  
  Indikator : Menyesuaikan dengan permasalahan dan tujuan yang ada

##Outlier

```python
def find_outliers(test_df2):
  hold = {}
  length = len(test_df2)
  for column in test_df2.columns[1:]:
    Q1 = test_df2[column].quantile(0.25)
    Q3 = test_df2[column].quantile(0.75)
    IQR = Q3 - Q1
    min = Q1 - 1.5*IQR
    max = Q3 + 1.5*IQR
    count = len(test_df2[(test_df2[column] < min) | (test_df2[column] > max)])
    hold[column] = {"count": count, "percent": count/length * 100}
  return pd.DataFrame(hold).transpose()

outlier_df = find_outliers(test_df2)
outlier_df
```
